```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(0)
```

```python
def generate_data(n_samples=100):
    citrus_acid = np.random.uniform(1, 5, n_samples)
    density = np.random.uniform(9930, 9970, n_samples)
    quality = 0.0015 * (citrus_acid / density) * 10  # Scale to fit the quality range
    return pd.DataFrame({'CitrusAcid': citrus_acid, 'Density': density, 'Quality': quality})
```

```python
data = generate_data()
```

```python
print(data.head())
```

```python
def assign_status(quality):
    if 5 <= quality < 6.5:
        return 'Normal'
    elif 6.5 <= quality < 8:
        return 'Good'
    elif 8 <= quality < 10:
        return 'Best'
    else:
        return 'Out of Range'

data['Status'] = data['Quality'].apply(assign_status)
```

```python
X = data[['CitrusAcid', 'Density']]
y = data['Quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.4f}")
```

```python
plt.figure(figsize=(14, 6))
```

```python
plt.subplot(1, 2, 1)
plt.scatter(X_test['CitrusAcid'], y_test, color='blue', label='Actual Quality', alpha=0.5)
plt.scatter(X_test['CitrusAcid'], y_pred, color='red', label='Predicted Quality', alpha=0.5)
plt.xlabel('Citrus Acid')
plt.ylabel('Quality')
plt.title('Citrus Acid vs Quality')
plt.legend()
```

```python
plt.subplot(1, 2, 2)
plt.scatter(X_test['Density'], y_test, color='blue', label='Actual Quality', alpha=0.5)
plt.scatter(X_test['Density'], y_pred, color='red', label='Predicted Quality', alpha=0.5)
plt.xlabel('Density')
plt.ylabel('Quality')
plt.title('Density vs Quality')
plt.legend()
```

```python
plt.tight_layout()
plt.show()
```
